```python
import tkinter as tk

class Page(tk.Frame):
    def __init__(self, *args, **kwargs):
        tk.Frame.__init__(self, *args, **kwargs)
    def show(self):
        self.lift()

class Page1(Page):
   def __init__(self, *args, **kwargs):
       Page.__init__(self, *args, **kwargs)
       label = tk.Label(self, text="Welcome\nto\nthe\nWarehouse", justify="center")
       label.config(font=("Courier",24))
       label.pack(side="top", expand=True)

class Page2(Page):
   def __init__(self, *args, **kwargs):
        Page.__init__(self, *args, **kwargs)
        label = tk.Label(self, text="Please enter the dimensions\nof the warehouse:")
        label.config(font=(20))
        label.pack(side="top", fill="both", expand=True)
        
        #for if it exists, winfo_exists()
#         list = root.grid_slaves()
#         for l in list:
#         l.destroy()



class Page3(Page):
   def __init__(self, *args, **kwargs):
       Page.__init__(self, *args, **kwargs)
       label = tk.Label(self, text="This is page 3")
       label.pack(side="top", fill="both", expand=True)

class MainView(tk.Frame):
    def __init__(self, *args, **kwargs):
        tk.Frame.__init__(self, *args, **kwargs)
        p1 = Page1(self) #welcome page
        p2 = Page2(self) #warehouse dimensions
        p3 = Page3(self)
 
        buttonframe = tk.Frame(self)
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        buttonframe.place(anchor="c", relx=.5, rely=.5)
        buttonframe.pack(side="top")
        #buttonframe.pack(side="top", fill="x", expand=False)
        bottomF = tk.Frame(self)
        bottomF.pack(side="bottom", pady=(20,20))

        p1.place(in_=container, x=0, y=0, relwidth=1, relheight=1)
        p2.place(in_=container, x=0, y=0, relwidth=1, relheight=1)
        p3.place(in_=container, x=0, y=0, relwidth=1, relheight=1)

        b1 = tk.Button(buttonframe, text="Continue", command=p2.lift) #continue button

#         b2 = tk.Button(buttonframe, text="Page 2", command=p2.lift)
#         b3 = tk.Button(buttonframe, text="Page 3", command=p3.lift)

        e1 = tk.Entry()
        e1.place(in_=p2)
        
        b1.pack(side="left")
#         b2.pack(side="left")
#         b3.pack(side="left")

        p1.show()

        
if __name__ == "__main__":
    root = tk.Tk()
    root.title("Quick-Space")
    main = MainView(root)
    main.pack(side="top", fill="both", expand=True)
    root.wm_geometry("300x400")
    root.mainloop()
```
